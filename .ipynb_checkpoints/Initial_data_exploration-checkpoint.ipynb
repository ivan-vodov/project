```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
import pandas as pd
```

```python
wages_file = "BLS_Data.csv"
wages_df = pd.read_csv(wages_file, thousands=',', low_memory=False)
wages_df.shape
```

```python
regr_param1, regr_param2, regr_param3 ='A_MEAN', 'A_Occ_Market', 'TOT_EMP'
regr_param1_prescaler, regr_param2_prescaler, regr_param3_prescaler=0.001,0.00001,0.01

#regr_param1, regr_param2, regr_param3 ='A_PCT25','A_MEAN','A_PCT75'
#regr_param1_prescaler, regr_param2_prescaler, regr_param3_prescaler=0.001,0.001,0.001
```

```python
wages_df['A_Occ_Market']=wages_df['A_MEAN']*wages_df['TOT_EMP']
wages_df.head()
```

```python
print(wages_df.shape)
wages_df.dropna(axis=0, how='any', subset=[regr_param1],inplace=True)
if(regr_param2!=''):
    wages_df.dropna(axis=0, how='any', subset=[regr_param2],inplace=True)
if(regr_param3!=''):
    wages_df.dropna(axis=0, how='any', subset=[regr_param3],inplace=True)
        
print(wages_df.shape)
```

```python
wages_df[wages_df['OCC_CODE']=='45-0000']
```

```python
prices_file = "Zillow_Sale_Prices.csv"
prices_df = pd.read_csv(prices_file)
prices_df.head()
```

```python
mapping_file = "BLS-Zillow-Mapping.csv"
mapping_df = pd.read_csv(mapping_file)
mapping_df.head()
```

```python
mapping_df=mapping_df.loc[:,['AREA','ZillowRegionID']]
mapping_df.head()
```

```python
prices_df=prices_df.loc[:,['ZillowRegionID','RegionName','2017-01']].rename(columns={'2017-01':'SalePrice'})
prices_df.head()
```

```python
data_consistency_check=wages_df.loc[:,['OCC_CODE','AREA','PRIM_STATE']].groupby(by=['OCC_CODE','AREA']).count()
data_consistency_check[data_consistency_check['PRIM_STATE']>1]
```

```python
regr_param1_df=wages_df.pivot(index='AREA', columns='OCC_CODE', values=regr_param1).reset_index()
regr_param1_df.dropna(axis=0,how='any',inplace=True)
regr_param1_df.iloc[:,1:]=regr_param1_df.iloc[:,1:].multiply(regr_param1_prescaler)
print(regr_param1_df.shape)
regr_param1_df.head()
```

```python
if(regr_param2!=''):
    regr_param2_df=wages_df.pivot(index='AREA', columns='OCC_CODE', values=regr_param2).reset_index()
    regr_param2_df.dropna(axis=0,how='any',inplace=True)
    regr_param2_df.iloc[:,1:]=regr_param2_df.iloc[:,1:].multiply(regr_param2_prescaler)
    print(regr_param2_df.shape)
```

```python
if(regr_param3!=''):
    regr_param3_df=wages_df.pivot(index='AREA', columns='OCC_CODE', values=regr_param3).reset_index()
    regr_param3_df.dropna(axis=0,how='any',inplace=True)
    regr_param3_df.iloc[:,1:]=regr_param3_df.iloc[:,1:].multiply(regr_param3_prescaler)
    print(regr_param3_df.shape)
```

```python
regr_params_df=regr_param1_df
if(regr_param2!=''):
    regr_params_df=regr_params_df.merge(regr_param2_df,on='AREA',how='inner', suffixes=('',regr_param2))
regr_params_df.shape
```

```python
if(regr_param3!=''):
    regr_params_df=regr_params_df.merge(regr_param3_df,on='AREA',how='inner', suffixes=('',regr_param3))
regr_params_df.shape
```

```python
regr_params_df.head()
```

```python
regr_params_df=regr_params_df.merge(mapping_df,on='AREA',how='inner')
regr_params_df.shape
```

```python
regr_params_df=regr_params_df.merge(prices_df, on='ZillowRegionID',how='inner')
regr_params_df.shape
```

```python
X_df=regr_params_df.drop(columns=['AREA','ZillowRegionID','RegionName','SalePrice'])
```

```python
fig, ax = plt.subplots(figsize=(20,7))
sb.heatmap(X_df.corr(),annot=False,center=0)
```

```python
from sklearn.preprocessing import scale
X=scale(X_df)
X.shape
```

```python
y=regr_params_df['SalePrice']
y.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# Fitting our model with all of our features in X
fit_results=model.fit(X, y)
predictions = model.predict(X)
```

```python
import matplotlib.lines as mlines
plt.axes().set_aspect('equal')
plt.scatter(y, predictions,alpha=0.5)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

print(f"R2 Score: {model.score(X,y)}")
print(f"MSE: {mean_squared_error(y,predictions)}")
print('Intercept:', model.intercept_)
```

```python
coeff_fd=pd.DataFrame(data={'param':X_df.columns.values,'coeff':model.coef_})
coeff_fd
```

# Dimensionality reduction

```python
from sklearn import decomposition
from sklearn.decomposition import PCA
```

```python
pca = decomposition.PCA()
model_pca = pca.fit_transform(X)

pd.DataFrame(data={'PC Var explained':pca.explained_variance_ratio_, 'cumulative Var explained': pd.Series(pca.explained_variance_ratio_).cumsum()}).head(15)
```

```python
#comps = pd.DataFrame(pca.components_, columns=list(X.columns.values))
comps = pd.DataFrame(pca.components_, columns=list(X_df.columns.values))
fig, ax = plt.subplots(figsize=(20,7))
sb.heatmap(comps,annot=False,center=0)
```

```python
from sklearn.feature_selection import RFECV
from sklearn.svm import SVR

estimator = SVR(kernel="linear")
selector = RFECV(estimator, step=1, cv=5)
selector = selector.fit(X, y)

selector_df=pd.DataFrame(data={'param':X_df.columns.values,'selected':selector.support_, 'rank': selector.ranking_})
print("Filtered out parameters:")
selector_df[selector_df['selected']==False]
```
